```python
import pygame
import random
import time
from datetime import datetime

# 1. 게임 초기화
pygame.init()

# 2. 게임창 옵션 설정
size = [400, 900]
screen = pygame.display.set_mode(size)

title = "My Game"
pygame.display.set_caption(title)

# 3. 게임 내 필요한 설정
clock = pygame.time.Clock()

class obj:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.move = 0  # 이동 속도 초기화
    def put_img(self, address):
        """이미지 주소를 받아서 해당 이미지를 로드합니다."""
        if address[-3:] == "png":
            self.img = pygame.image.load(address).convert_alpha()  # PNG 이미지 로드 (투명도 지원)
        else:
            self.img = pygame.image.load(address)  # 캐릭터 이미지 로드 (예시)
            self.sx, self.sy = self.img.get_size()  # 이미지 크기 가져오기
    def change_size(self, sx, sy):
        """이미지 크기를 변경합니다."""
        self.img = pygame.transform.scale(self.img, (sx, sy))
        self.sx, self.sy = self.img.get_size()
    def show(self):
        """이미지를 화면에 그립니다."""
        screen.blit(self.img, (self.x, self.y))

def crash(a, b):
    """두 객체가 충돌하는지 확인합니다."""      
    if (a.x - b.sx <= b.x) and (b.x <= a.x + a.sx):
        if (a.y - b.sy <= b.y) and (b.y <= a.y + a.sy):
            return True
        else:
            return False
    else:
        return False
    

ss = obj()  # 캐릭터 객체 생성
ss.put_img("C:/Users/eggmoneynah/Desktop/증명사진_신인택.jpg")
ss.change_size(50, 80)  # 이미지 크기 조정 (예시)
ss.x = round(size[0]/2 - ss.sx/2)  # 캐릭터 위치 설정 
ss.y = size[1] - ss.sy - 20  # 캐릭터 위치 설정
ss.move = 10  # 이동 속도 설정

left_go = False  # 왼쪽 이동 초기화
right_go = False  # 오른쪽 이동 초기화
space_go = False  # 스페이스바 초기화 (예시로 사용)

m_list = [] # 미사일 리스트 초기화
a_list = []  # 적 미사일 리스트 초기화

black = (0, 0, 0)  # 배경색 설정 (검정색)
white = (255, 255, 255)  # 흰색 설정
k = 0  # 키 입력 초기화

GO = 0  # 게임 종료 변수 초기화
kill = 0  # 적군 제거 카운트 초기화
loss = 0 

# 4- 0. 게임 시작 대기 화면
SB = 0  # 게임 시작 대기 변수 초기화
while SB == 0:
    clock.tick(60)  # 초당 60프레임 설정
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                SB = 1  # 스페이스바가 눌리면 게임 시작
    screen.fill(black)  # 배경색으로 화면 채우기
    font = pygame.font.Font("C:/Windows/Fonts/ariblk.ttf", 15)  # 폰트 설정 (글씨체, 크기)
    text = font.render("PRESS SAPCE KEY TO START THE GAME".format(kill, loss), True, (255,255,255)) # 텍스트 렌더링 하얀색으로 (내용, 색깔)
    screen.blit(text, (30, round(size[1]/2 - 50)))  # 텍스트 위치 설정 (x, y)
    pygame.display.flip()



# 4. 메인 이벤트

start_time = datetime.now()  # 현재 시간 가져오기
SB = 0
while SB == 0:
    
    # 4-1. FPS 설정
    clock.tick(60)  # 초당 60프레임 설정

    # 4-2. 각종 입력 감지
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            SB = 1
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                left_go = True
            elif event.key == pygame.K_RIGHT:
                right_go = True
            elif event.key == pygame.K_SPACE:
                space_go = True
                k = 0  # 스페이스바가 눌렸을 때 키 입력 카운트 초기화
        elif event.type ==pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_go = False
            elif event.key == pygame.K_RIGHT:
                right_go = False
            elif event.key == pygame.K_SPACE:
                space_go = False

        # elif event.type == pygame.KEYDOWN:
        #     if event.key == pygame.K_ESCAPE:
        #         SB = 1  

    # 4-3. 입력, 시간에 따른 변화
    now_time = datetime.now()  # 현재 시간 업데이트
    delta_time = round((now_time - start_time).total_seconds())  # 시작 시간과 현재 시간의 차이 계산

    if left_go == True:
        ss.x -= ss.move  # 왼쪽 이동
        if ss.x < 0:  # 화면 밖으로 나가지 않도록 제한
            ss.x = 0
    elif right_go == True:
        ss.x += ss.move # 오른쪽 이동
        if ss.x > size[0] - ss.sx:  # 화면 밖으로 나가지 않도록 제한
            ss.x = size[0] - ss.sx
    

    if space_go == True and k % 6 == 0:  # 스페이스바가 눌렸을 때 그리고 k 나누기 6이 나누어 떨어질때 미사일 발사
        mm = obj()  # 캐릭터 객체 생성
        mm.put_img("C:/Users/eggmoneynah/Desktop/증명사진_신인택.jpg")
        mm.change_size(5, 15)  # 이미지 크기 조정 (예시)
        mm.x = round(ss.x + ss.sx/2 - mm.sx/2)
        mm.y = ss.y - mm.sy - 10
        mm.move = 20  # 이동 속도 설정
        m_list.append(mm)
    k += 1  # 키 입력 카운트 증가
    d_list = []
    for i in range(len(m_list)):
        m = m_list[i]
        m.y -= m.move  # 미사일 이동
        if m.y <= -m.sy:
            d_list.append(i)
    d_list.reverse()  # 역순으로 삭제하기 위해 리스트를 뒤집음
    for d in d_list:
        del m_list[d]

    if random.random() > 0.98:  # 2% 확률로 미사일 생성
        aa = obj()  # 캐릭터 객체 생성
        aa.put_img("C:/Users/eggmoneynah/Desktop/증명사진_신인택.jpg")
        aa.change_size(40, 40)  # 이미지 크기 조정 (예시)
        aa.x = random.randrange(0, size[0] - aa.sx - round(ss.sx/2))  # 랜덤 위치 설정
        aa.y = 10
        aa.move = 1  # 이동 속도 설정
        a_list.append(aa)
    d_list = []    
    for i in range(len(a_list)):
        a = a_list[i]
        a.y += a.move  # 적군 이동
        if a.y >= size[1]:
            d_list.append(i)
    d_list.reverse()  # 역순으로 삭제하기 위해 리스트를 뒤집음
    for d in d_list:
        del a_list[d]
        loss += 1  # 적군이 화면 밖으로 나가면 카운트 증가

    # 4-3-2. 미사일과 적군 충돌 확인
    # 충돌 확인을 위해 두 리스트를 비교 
    dm_list = []  # 미사일 삭제 리스트 초기화
    da_list = []  # 적군 삭제 리스트 초기화
    for i in range(len(m_list)):
        for j in range(len(a_list)):
            m = m_list[i]
            a = a_list[j]
            if crash(m, a) == True :  # 미사일과 적군 충돌 확인
                dm_list.append(i)
                da_list.append(j)
    dm_list = list(set(dm_list))  # 중복 제거
    da_list = list(set(da_list))  # 중복 제거

    for dm in dm_list:
        del m_list[dm]
    for da in da_list:
        del a_list[da]
        kill += 1

    # dm_list.reverse()  # 역순으로 삭제하기 위해 리스트를 뒤집음
    # da_list.reverse()  # 역순으로 삭제하기 위해 리스트를 뒤집음

    # 4-3-3. 적군과 캐릭터 충돌 확인
    for i in range(len(a_list)):
        a = a_list[i]
        if crash(a, ss) == True:
            SB = 1
            GO = 1


    # 4-4. 그리기
    screen.fill(black)  # 배경색으로 화면 채우기
    ss.show()  # 캐릭터 이미지 그리기 (예시 위치)
    for m in m_list:
        m.show()    # 미사일 이미지 그리기
    for a in a_list:
        a.show()    # 적군 이미지 그리기

    font = pygame.font.Font("C:/Windows/Fonts/ariblk.ttf", 20)  # 폰트 설정 (글씨체, 크기)
    text_kill = font.render("killed : {}  loss : {}".format(kill, loss), True, (255,255,0)) # 텍스트 렌더링 노란색으로 (내용, 색깔)
    screen.blit(text_kill, (10, 10))  # 텍스트 위치 설정 (x, y)

    text_time = font.render("time : {}".format(delta_time), True, (255,255,255)) # 텍스트 렌더링 하얀색으로 (내용, 색깔)
    screen.blit(text_time, (size[0] - 100, 10))  # 텍스트 위치 설정 (x, y)

    # 4-5. 화면 업데이트
    pygame.display.flip()

# 5. 게임 종료
while GO == 1:
    clock.tick(60)  # 초당 60프레임 설정
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            GO = 0
    font = pygame.font.Font("C:/Windows/Fonts/ariblk.ttf", 40)  # 폰트 설정 (글씨체, 크기)
    text = font.render("GAME OVER".format(kill, loss), True, (255,0,0)) # 텍스트 렌더링 빨간색으로 (내용, 색깔)
    screen.blit(text, (60, round(size[1]/2 - 50)))  # 텍스트 위치 설정 (x, y)
    pygame.display.flip()

pygame.quit()
# exit(0)
```
